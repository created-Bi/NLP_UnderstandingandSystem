````python
from random import choice
# 生成文法对应的字典 unstructural data -> structural data
def create_grammar(grammar_str, split='->', line_split='\n'):
    grammar = {}
    for i, line in enumerate(grammar_str.split(line_split)):
        #print('the {}th line: {}'.format(i, line))
        if not line.strip(): continue
        left, right = line.split(split)
        grammar[left.strip()] = [tup.split() for tup in right.split('|')]
    return grammar
# 根据文法字典生成句子
def generate(grammar, target):
    if target not in grammar: return target
    
    expanded = [generate(grammar, t) for t in choice(grammar[target])]
    return ''.join([e if e != '/n' else '\n' for e in expanded if e != 'null'])

'''
    :param: n, the number of the sentences you want to generate
    :param: start, the start of the grammar
'''
def generate_n(n, start, gram_str):
    grammar = create_grammar(gram_str, split='=')
    sentences = []
    for i in range(n):
        sentences.append(generate(grammar, start))
    return sentences
````

````python
TOKEN_BIGRAM = [''.join(TOKEN[i:i+2]) for i in range(len(TOKEN[:-2]))]
words_count_bigram = Counter(TOKEN_BIGRAM)
print(words_count_bigram.most_common(20))
````

````python
def prob_2(word1, word2):  # p(w1,w2) = count(w1,2)/count(w1)
    if word1 + word2 in words_count_bigram: return words_count_bigram[word1+word2] / words_count[word1]
    else:
        return 1 / len(words_count_bigram)
    
def get_probablity(sentence):
    words = cut(sentence)
    
    sentence_pro = 1
    length = len(words)
    
    for i, word in enumerate(words[:-1]):
        next_ = words[i+1]
        
        probability = prob_2(word, next_)  # p(w1|w2)
        
        sentence_pro *= probability  # p(s) = p(w1)*p(w2|w1)*p(w3|w2)..p(wn|wn-1) 
        sentence_pro = sentence_pro ** (1/length)
        
    return sentence_pro
````
