## 机器学习的三个基本要素
### 1. 模型：
由于我们不知道真实的映射函数 $ g(x) $的具体形式，因而只能根据经验来假设一个函数集合$ F $，称为假设空间，然后通过观测其在训练集$D$，从中选择一个理想的假设。
假设空间$F$通常为一个参数化的函数族：
$$ F={\{f(x;\theta)|\theta ∈ R^D\}} $$
其中$f(x;\theta)$是参数为$\theta$的函数，也成为模型，$D$为参数的数量。

### 2.学习准则：
令训练集$D$由独立同分布的样本组成，一个好的模型就是使：
$$ |f(x,\theta^*) - y)| < min $$

说白了，就是设置损失函数，利用损失来量化两个变量之间的差异。

$$ R(\theta) = E_{(x,y), y)}[L(y,f(x;\theta))] $$

### 3.损失函数：
1. 0-1损失函数；
2. 平方损失函数；
3. 交叉熵损失函数等

## 1.1 用 numpy 实现一个基于（随机）梯度下降的线性回归

```python
learning_rate = 0.0000001
epoch = 10000
min_loss = float('inf')
best_w0, best_w1 = 0, 0
```

```python
w0, w1 = rd.normal(), rd.normal()

for e in range(epoch):

    y_hat = w1 * x + w0 # (200, )
    #print("y_hat = {} * x + {}".format(w1, w0))
    loss = 0.5 * np.mean((y_with_noise - y_hat) ** 2) # (1, ) 
    
    d_w0 = (-1) * np.mean((y_with_noise - y_hat) * 1)
    d_w1 = (-1) * np.mean((y_with_noise - y_hat) * x)
    
    w0 = w0 + (-1) * d_w0 * learning_rate
    w1 = w1 + (-1) * d_w1 * learning_rate
    
    if loss < min_loss:
        min_loss = loss
        best_w0, best_w1 = w0, w1
        print("Loss updated to {} at the {}th epoch.".format(loss, e))
```

```python
plt.scatter(x, y_with_noise)
plt.plot(x, best_w1 * x + best_w0, color='red')
print(best_w0, best_w1)
```

## 1.2 用 numpy 实现一个解析解线性回归

$$ ▽_{θ}J(θ) = X^{T}Xθ - X^{T}y $$

### 为求得极值点，需令上式等式等于0 ==》

$$  X^{T}Xθ = X^{T}y $$

$$ θ = (X^{T}X)^{-1}X^{T}y $$

### 对于一维输入 X 来说， 令 k = y/x 即可 ==》

```python
w = np.mean(y_with_noise / x)
```

```python
w
```

```python
b = np.mean(y_with_noise - w * x)
```

```python
b
```

```python
plt.scatter(x, y_with_noise)
plt.plot(x, w * x + b, color='red')
```

```python
X_train, X_test = X[test_num:], X[:test_num]
y_train, y_test = y[test_num:], y[:test_num]
```

```python
learning_rate = 0.0000001
epoch = 20000
min_loss = float('inf')
best_w0, best_w1 = 0, 0
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-1*x))

def cross_entropy(y, y_hat):
    return -1 * (y * np.log(y_hat) + (1-y) * np.log(1-y_hat))
```

```python
K, b = 0.1, 0.5
```

```python
y = K * X + b
```

估计函数：$$ \hat{y} = sigmoid(WX + b) $$

损失函数：$$ Loss = -(y\log{\hat{y}} + (1-y)\log{(1-\hat{y})}) $$

$$ \frac{\delta{Loss}}{\delta{W}} = \frac{\delta{Loss}}{\delta{\hat{y}}} \frac{\delta{\hat{y}}}{\delta{sigmoid}} \frac{\delta{sigmoid}}{\delta{W}} $$

$$ Multi-class CrossEntropy L = -\sum_i{y_i * \log{\hat{y_i}}} $$

## Softmax函数
$$ softmax = \frac{e^x}{\sum_i{e_i^x}} $$

## 数据预处理

### 作用：让我们处理的数据标准化（更好被拟合）

### 1. 线性标准化X => min, max

$$ X' = \frac{X-min(X)}{max(X) - min(X)} $$

### 2. X => $均值 \mu $ =>  方差 

$$ X' = \frac{X-\mu}{方差} $$

```python
mean_x = np.mean(X_train, axis=0)
std_x = np.std(X_train, axis=0)
```

```python
X_train = (X_train - mean_x) / std_x
```

```python
w0, w1 = np.zeros(1), rd.normal(size=(30,)) * 0.0001
# print(np.dot(X, w1) + w0)
for e in range(epoch):
    
    y_hat = sigmoid(np.dot(X_train, w1) + w0) # (569, )
    #print(y_hat)
    loss = np.mean(cross_entropy(y_train, y_hat))
    #print(loss)
    d_w0 = np.mean(y_hat - y_train)
    d_w1 = np.dot(X_train.T, (y_hat - y_train))
    
    w0 = w0 + (-1) * d_w0 * learning_rate
    w1 = w1 + (-1) * d_w1 * learning_rate
    
    if loss < min_loss:
        min_loss = loss
        best_w0, best_w1 = w0, w1
        print("Loss updated to {} at the {}th epoch.".format(loss, e))
    
```
